# Create column 'Average_Score'

```python
# create binary columns for the calculation of 'Average_Score'
def binary_cols(df):
    for x in df.iloc[:,4:]:
        df[x + str('_bin')] = np.where(df[x]>=3,1,0)
    return df
```

```python
# WS
df_binary = df_ws.copy()
df_ws_binary = binary_cols(df_binary)
df_ws_binary['Average_Score'] = df_ws_binary.iloc[:,4:].mean(axis=1)
WS = pd.concat([df_ws,df_ws_binary.iloc[:,4:]],axis=1)
```

```python
# BA
df_binary = df_ba.copy()
df_ba_binary = binary_cols(df_binary)
df_ba_binary['Average_Score'] = df_ba_binary.iloc[:,4:].mean(axis=1)
BA = pd.concat([df_ba,df_ba_binary.iloc[:,4:]],axis=1)
```

```python
# TR
df_binary = df_tr.copy()
df_tr_binary = binary_cols(df_binary)
df_tr_binary['Average_Score'] = df_tr_binary.iloc[:,4:].mean(axis=1)
TR = pd.concat([df_tr,df_tr_binary.iloc[:,4:]],axis=1)
```
